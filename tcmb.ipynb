```python
import requests
import pandas as pd

url = "https://evds2.tcmb.gov.tr/EVDSServlet"

headers = {
    'Accept': '*/*',
    'Accept-Language': 'tr-TR,tr;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
    'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8',
    'Origin': 'https://evds2.tcmb.gov.tr',
    'Referer': 'https://evds2.tcmb.gov.tr/index.php?/evds/serieMarket',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest'
}

s = "01-11-2023"
e = "02-02-2024"

sd = pd.to_datetime(s)
ed = pd.to_datetime(e)

take = ed - sd
days = take.days

i = int((days/20))
print(f"toplam {i+1} set")
r = days % 20 

dfs = []

tickers = [
    "TP.DK.USD.A.YTL-0",
    "TP.MK.F.BILESIK-0"
]

for ticker in tickers:

    for x in range(0,i):
      skip = x*20
      take = (x+1)*20

      print(f"{ticker} set {x+1}")
    
      data = {
          'orderby': 'Tarih+desc',
          'thousand': '1',
          'decimal': '2',
          'frequency': 'WORKDAY',
          'aggregationType': 'avg',
          'formula': '0',
          'graphicType': '0',
          'skip': skip,
          'take': take,
          'sort': 'Tarih#true',
          'select': ticker,
          'startDate': s,
          'endDate': e,
          'obsCountEnabled': '',
          'obsCount': '',
          'categories': '5863',
          'mongoAdresses': 'evds',
          'userId': '',
          'datagroupString': 'bie_dkdovytl',
          'dateFormatValue': 'dd-mm-yyyy',
          'customFormula': 'null',
          'excludedSeries': 'null'
      }

      response = requests.post(url, headers=headers, data=data)

      json_data = response.json()

      df = pd.DataFrame(json_data["items"])
      dfs.append(df)

    if r > 0:
      print(f"{ticker} set {i+1}")
      skip = i*20
      take = (i+1)*20
      data = {
        'orderby': 'Tarih+desc',
        'thousand': '1',
        'decimal': '2',
        'frequency': 'WORKDAY',
        'aggregationType': 'avg',
        'formula': '0',
        'graphicType': '0',
        'skip': skip,
        'take': take,
        'sort': 'Tarih#true',
        'select': ticker,
        'startDate': s,
        'endDate': e,
        'obsCountEnabled': '',
        'obsCount': '',
        'categories': '5863',
        'mongoAdresses': 'evds',
        'userId': '',
        'datagroupString': 'bie_dkdovytl',
        'dateFormatValue': 'dd-mm-yyyy',
        'customFormula': 'null',
        'excludedSeries': 'null'
      }

      response = requests.post(url, headers=headers, data=data)

      json_data = response.json()

      df = pd.DataFrame(json_data["items"])
      dfs.append(df)

df = pd.concat(dfs) #index üstünden birleştirmek gerek
df.index = pd.to_datetime(df["Tarih"], format="%d-%m-%Y")
df = df.drop(columns=["Tarih", "UNIXTIME"])
df = df.sort_index()
df.to_csv("tcmb_data.csv")
print("tcmb_data.csv kaydedildi")
```
